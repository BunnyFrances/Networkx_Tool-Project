```python
#This is a simple example of showing the network(relationship) between different fruits.
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Create an empty Graph called G
G=nx.Graph()
```

```python
#add edges to graph G
G.add_edge('Apple', 'Mango') 
G.add_edge('Mango', 'Orange') 
G.add_edge('Grape', 'Banana') 
G.add_edge('Banana', 'Apple') 
G.add_edge('Orange', 'Plum') 
G.add_edge('Pinapple', 'Plum')
G.add_edge('Mango', 'Plum')
G.add_edge('Pinapple', 'Orange')
```

```python
# add nodes to graph with color=red
G.add_nodes_from(['Apple', 'Banana','Orange','Grape','Mango', 'Pinapple','Plum'])
```

```python
#We want the graph layouts to be shell shape
graph_shape = nx.shell_layout(G)
```

```python
#we add blue nodes into graph G. node_color controls for the color of each node, and alpha controls the darkness of that color
nx.draw_networkx_nodes(G, graph_shape, node_size=800, node_color=['red','blue','yellow'], alpha=0.2)
plt.show()
```

```python
#we add network edges into graph G
nx.draw_networkx_edges(G, graph_shape)
plt.show()
```

```python
#we add labels to graph G, and you can adjust the font and its size.
nx.draw_networkx_labels(G, graph_shape, font_size=12, font_family='sans-serif')
plt.show()
```

```python
#run the above 3 stes together and show the network about relationship between fruits.
nx.draw_networkx_nodes(G, graph_shape, node_size=1000, node_color=['red','blue','yellow'], alpha=0.3)
nx.draw_networkx_edges(G, graph_shape)
nx.draw_networkx_labels(G, graph_shape, font_size=10, font_family='sans-serif')

plt.show()
```
